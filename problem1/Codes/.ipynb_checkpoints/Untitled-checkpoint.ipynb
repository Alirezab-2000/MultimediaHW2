ابتدا باید موارد مورد نیاز را ایمپورت کنیم

```python
import numpy
import cv2
import matplotlib.pyplot as plt
import cmath as cmath
```

حال تابعی تعریف میکنیم تا با استفاده از آن مقادیر W را محاسبه کنیم

```python
def calculateW(u:int , v:int):
    basicImage = numpy.array([])
    
    for x in range(8):
        row = numpy.array([])
        for y in range(8):
            expValue = cmath.exp(((-1)**0.5) * 2 * numpy.pi * ((u * x + v * y)/8));
            row = numpy.append(row , expValue)
        
        basicImage = numpy.append(basicImage , row) if basicImage.size == 0 else numpy.vstack((basicImage , row))
        
    return basicImage
    
```

حال با زدن حلقه میتوان کل ماتریس حاصل از تصاویر پایه را محاسبه کرد


```python
transformMatrix = numpy.array([])
for u in range(8):
    transformMatrixRow = numpy.array([])
    for v in range(8):
        basicImage = calculateW(u , v)
        if len(transformMatrixRow) == 0:
            transformMatrixRow = numpy.array( [basicImage])
        else:
            transformMatrixRow = numpy.append(transformMatrixRow,[basicImage],axis=0)
    if len(transformMatrix) == 0:
        transformMatrix = numpy.array([transformMatrixRow])
    else:
        transformMatrix = numpy.vstack((transformMatrix , [transformMatrixRow]))
```

حال این تصاویر پایه را نشان میدهیم

```python
fig , subplot = plt.subplots(8 , 8 , figsize=(17, 17))

for i in range(8):
    for j in range(8):
        subplot[i][j].imshow(transformMatrix[i][j].real , cmap = "gray")
        subplot[i][j].set_title(""+str(i)+" "+str(j))

plt.show()
```

و همچنین تصویر سمت راست به دلیل داشتن ساختار مربعی، در تبدیل فوریه مربوط به آن تصاویر مربوط به مرکز که بیشتر ساختار هایی با فرکانس تغییراات بالا را مشخص میکنند، باید وزن بیشتری در ساختن این تصویر داشته باشند.




پ) برای حل قسمت سوم سوال دو تصویر را میخوانیم و تبدیل فوریه آن ها را حساب میکنیم


```python
verticalImage = cv2.imread("../Vertical.png",cv2.IMREAD_GRAYSCALE)
boardImage = cv2.imread("../CheckerBoard.png" , cv2.IMREAD_GRAYSCALE)

fftVerticalImage = numpy.fft.fft2(verticalImage)
fftBoardImage = numpy.fft.fft2(boardImage)
```

حال برای تصویر vertical بخش ها موهومی و واقعی و فاز و اندازه را نشان میدهیم

```python
realVerticalImage = fftVerticalImage.real
imagVerticalImage = fftVerticalImage.imag
absVerticalImage = numpy.log(1 + numpy.abs(fftVerticalImage))
phaseVerticalImage = numpy.arctan2(imagVerticalImage , realVerticalImage)

fig2 , subplot2 = plt.subplots(2 , 2 , figsize=(15, 15))

subplot2[0][0].imshow(realVerticalImage , cmap = "gray")
subplot2[0][0].set_title("real")

subplot2[0][1].imshow(imagVerticalImage , cmap = "gray")
subplot2[0][1].set_title("imag")

subplot2[1][0].imshow(absVerticalImage, cmap = "gray")
subplot2[1][0].set_title("abs")

subplot2[1][1].imshow(phaseVerticalImage, cmap = "gray")
subplot2[1][1].set_title("phase")

plt.show()
```

حال برای تصویر جدولی همان مقادیر را نشان میدهیم

```python
realBoardImage = fftBoardImage.real
imagBoardImage = fftBoardImage.imag
absBoardImage = numpy.log(1 + numpy.abs(fftBoardImage))
phaseBoardImage = numpy.arctan2(imagBoardImage , realBoardImage)

fig2 , subplot2 = plt.subplots(2 , 2 , figsize=(15, 15))

subplot2[0][0].imshow(realBoardImage , cmap = "gray")
subplot2[0][0].set_title("real")

subplot2[0][1].imshow(imagBoardImage , cmap = "gray")
subplot2[0][1].set_title("imag")

subplot2[1][0].imshow(absBoardImage, cmap = "gray")
subplot2[1][0].set_title("abs")

subplot2[1][1].imshow(phaseBoardImage, cmap = "gray")
subplot2[1][1].set_title("phase")

plt.show()
```
