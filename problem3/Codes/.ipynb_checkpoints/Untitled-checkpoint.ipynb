```python
import numpy
import cv2
import matplotlib.pyplot as plt
```

ابتدا تابعی مینویسیم که یک مربع سفید در دل مربع سیاه ایجاد کند

```python
def createRect (blackSize:int , whiteSize:int , center:int = 0):
    if center == 0:
        center = blackSize / 2
    
    Rect = numpy.array([[1 if ((numpy.abs(x - center) <= ( whiteSize / 2)) and (numpy.abs(y - center) <= (whiteSize / 2))) else 0 for x in range(blackSize)] for y in range(blackSize)])
    return Rect
```

حال یک مربع میسازیم که مربع سفید 10 در 10 در وسط یک مربع 100 در 100 باشد.

```python
firstRect = createRect(100 , 10)
firstRect_freq = numpy.fft.fft2(firstRect)
firstRect_freq_shift = numpy.fft.fftshift(firstRect_freq)

# print(firstRect)
fig2 , subplot = plt.subplots(1 , 2 , figsize=(15, 15))
subplot[0].imshow(firstRect , cmap = "gray")
subplot[0].set_title("special mode")

subplot[1].imshow(numpy.log(1 + numpy.abs(firstRect_freq_shift)) , cmap = "gray")
subplot[1].set_title("shift frequency mode")
```

مربع را حرکت میدهیم

```python
secondRect = createRect(100 , 10 , 25)
secondRect_freq = numpy.fft.fft2(secondRect)
secondRect_freq_shift = numpy.fft.fftshift(secondRect_freq)

# print(firstRect)
fig2 , subplot = plt.subplots(1 , 2 , figsize=(15, 15))
subplot[0].imshow(secondRect , cmap = "gray")
subplot[0].set_title("special mode")

subplot[1].imshow(numpy.log(1 + numpy.abs(secondRect_freq_shift)) , cmap = "gray")
subplot[1].set_title("shift frequency mode")
```
