موارد مورد نیاز را داخل پروژه ایمپورت میکنیم

```python
import numpy
import cv2
import matplotlib.pyplot as plt
```

ابتدا تصویر لنا را میخوانیم و فیلتر های سوبل در جهت عمودی و افقی را تعریف میکنیم و از آن ها را تبدیل فوریه میگیریم. 

```python
sobelY = numpy.array([[-1, -2, -1],[0,0,0],[1, 2 , 1]])
sobelX = numpy.array([[-1, 0, 1],[-2,0,2],[-1, 0 , 1]])
lena = cv2.imread("../lena512.bmp",cv2.IMREAD_GRAYSCALE)
shapesize = lena.shape[0]

# lena frequency
lena_freq = numpy.fft.fft2(lena)

# sobel frequency
sobelY_padded = numpy.zeros([shapesize, shapesize])
sobelX_padded = numpy.zeros([shapesize, shapesize])

sobelY_padded[0:(sobelY.shape[0]),0:(sobelY.shape[1])] = sobelY[:,:]
sobelX_padded[0:(sobelX.shape[0]),0:(sobelX.shape[1])] = sobelX[:,:]

sobelY_freq = numpy.fft.fft2(sobelY_padded)
sobelX_freq = numpy.fft.fft2(sobelX_padded)
```

حال فیلتر های سوبل را بر روی تصویر لنا در فاز مکان زمان اعمال میکنیم.

```python
special_lena_filtred_Y = cv2.filter2D(lena , -1 , sobelY)
special_lena_filtred_X = cv2.filter2D(lena , -1 , sobelX)

fig2 , subplot = plt.subplots(1 , 2 , figsize=(15, 15))
subplot[0].imshow(special_lena_filtred_Y , cmap = "gray")
subplot[0].set_title("filtered lena in special Y axis")

subplot[1].imshow(special_lena_filtred_X , cmap = "gray")
subplot[1].set_title("filtered lena in special X axis")
```

در مرحله بعد این فیلتر ها را در حوزه فرکانس اعمال میکنیم و نتیجه را نمایش میدهیم.

```python
frequnce_lena_filtred_Y = lena_freq * sobelY_freq
frequnce_lena_filtred_X = lena_freq * sobelX_freq

lena_filtred_Y = numpy.fft.ifft2(frequnce_lena_filtred_Y)
lena_filtred_X = numpy.fft.ifft2(frequnce_lena_filtred_X)

fig2 , subplot = plt.subplots(1 , 2 , figsize=(15, 15))
subplot[0].imshow(numpy.abs(lena_filtred_Y).astype(float) , cmap = "gray")
subplot[0].set_title("filtered lena in frequnce Y axis")

subplot[1].imshow(numpy.abs(lena_filtred_X).astype(float) , cmap = "gray")
subplot[1].set_title("filtered lena in frequnce X axis")
```

با مقایسه تصاویر بالا (چهار تصویر) میبینیم که تغییر زیادی بین آن ها مشاهده نمیشود و نتایج یکی است.
